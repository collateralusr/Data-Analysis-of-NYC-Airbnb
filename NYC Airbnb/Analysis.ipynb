# Data Analysis of NYC Airbnb 


```python
# Importing necessary libraries for future analysis of the dataset


import numpy as np  # NumPy is used for numerical operations and array manipulation.
import pandas as pd  # Pandas is used for data manipulation and analysis.
import matplotlib.pyplot as plt  # Matplotlib is a plotting library for creating visualizations.
import matplotlib.image as mpimg  # Matplotlib's image module for working with images.
%matplotlib inline  
import seaborn as sns  # Seaborn is a data visualization library based on Matplotlib
```

```python
# Using pandas library and 'read_csv' function to read NYC csv file.

# The file is already formatted for us from Kaggle.
airbnb = pd.read_csv("AB_NYC_2019.csv")
# Examining the first 10 rows of the DataFrame to understand its structure and contents.
airbnb.head(10)
```

•	id: Unique identifier for each Airbnb listing.

•	name: Descriptive name or title of the listing.

•	host_id: Unique identifier for the host of the listing.

•	host_name: Name of the host.

•	neighbourhood_group: The borough (group of neighborhoods) where the listing is located.

•	neighbourhood: The specific neighborhood where the listing is located.

•	latitude: Latitude coordinate of the listing location.

•	longitude: Longitude coordinate of the listing location.

•	room_type: Type of room offered in the listing (e.g., Entire home/apt, Private room, Shared room).

•	price: Price per night for renting the listing.

•	minimum_nights: Minimum number of nights required for booking.

•	number_of_reviews: Total number of reviews the listing has received.

•	last_review: Date of the last review received for the listing.

•	reviews_per_month: Average number of reviews per month for the listing.

•	calculated_host_listings_count: Number of listings the host has on Airbnb.

•	availability_365: Number of days the listing is available for booking within the next 365 days.

# Data Cleaning 


```python

# Check for missing values
airbnb.isnull().sum()

# Handle missing values appropriately 

# Check for duplicates
print("Number of duplicate rows:", airbnb.duplicated().sum())

# Drop duplicates if necessary
airbnb = airbnb.drop_duplicates()

# Convert data types
airbnb['last_review'] = pd.to_datetime(airbnb['last_review'])

# Handle categorical variables if needed 
```

# Exploratory Data Analysis (EDA)


```python
# Visualize distribution of numerical variables
sns.histplot(airbnb['price'], bins=30, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Explore correlations between numerical variables
sns.heatmap(airbnb.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Visualize categorical variables
sns.countplot(x='neighbourhood_group', data=airbnb)
plt.title('Number of Listings by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.show()
```

•	Price Distribution Summary:

    o	Mean Price: $152.72.

    o	Median Price: $106.0.

    o	Standard Deviation: $240.15, indicating a wide variability in prices.

    o	Minimum Price: The minimum price for a listing is $0.

    o	Maximum Price: The maximum price for a listing is $10,000.

•	Correlation Matrix:

    o	There is a weak positive correlation between price and longitude, indicating that as longitude increases, price  tends to slightly decrease.

    o	There is a moderate positive correlation between number of reviews and reviews per month, suggesting that listings with more reviews tend to have a higher number of reviews per month.

    o	There is a weak positive correlation between price and availability 365, indicating that as availability increases, price tends to slightly increase.

•	Number of Listings by Neighborhood Group:

    o	Manhattan has the highest number of listings, with 21,661 listings.

    o	Brooklyn follows closely behind with 20,104 listings.

    o	Queens, Bronx, and Staten Island have fewer listings compared to Manhattan and Brooklyn, with 5,666, 1,091, and 373 listings respectively.



# Room Type Distribution:


```python
# Visualize distribution of room types
sns.countplot(x='room_type', data=airbnb)
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()

# Explore room type distribution as percentages
room_type_percentage = airbnb['room_type'].value_counts(normalize=True) * 100
print("Room Type Distribution (%):")
print(room_type_percentage)
```

# Room Type Distribution:

```python
# Identify top hosts with the most listings
top_hosts = airbnb['host_id'].value_counts().head(10)
print("Top 10 Hosts with the Most Listings:")
print(top_hosts)

# Analyze distribution of host listings count
host_listings_count_distribution = airbnb['calculated_host_listings_count'].value_counts()
print("Distribution of Host Listings Count:")
print(host_listings_count_distribution)

# Investigate relationship between host characteristics and listing attributes
host_stats = airbnb.groupby('host_id').agg({
    'price': 'mean',
    'number_of_reviews': 'sum',
})
print("Host Statistics:")
print(host_stats.head())

# Room Type Distribution as percentages
room_type_percentage = airbnb['room_type'].value_counts(normalize=True) * 100
print("Room Type Distribution (%):")
print(room_type_percentage)
```

•	Entire home/apartment: Represents the largest proportion of listings, accounting for approximately 52% of all listings.

•	Private room: Represents a substantial portion of listings, comprising around 46% of all listings.

•	Shared room: Represents the smallest proportion of listings, making up only about 2% of all listings.


# Host Analysis:

```python
# airbnb['last_review'] = pd.to_datetime(airbnb['last_review'])

# Explore distribution of listings over time 
airbnb['last_review_month'] = airbnb['last_review'].dt.month
sns.countplot(x='last_review_month', data=airbnb)
plt.title('Distribution of Listings by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()

# Analyze trends in availability and prices over time
availability_trend = airbnb.groupby('last_review').mean(numeric_only=True)['availability_365']
availability_trend.plot()
plt.title('Trend of Availability Over Time')
plt.xlabel('Date')
plt.ylabel('Availability (days)')
plt.show()


# Investigate seasonal patterns in demand and prices
airbnb['last_review_month'] = airbnb['last_review'].dt.month
monthly_price_median = airbnb.groupby('last_review_month')['price'].median()
monthly_price_median.plot()
plt.title('Seasonal Pattern in Prices')
plt.xlabel('Month')
plt.ylabel('Median Price')
plt.show()
```

•	Top 10 Hosts with the Most Listings:

    o	The host with host_id 219517861 has the highest number of listings, with 327 listings.

    o	The second highest host has host_id 107434423 with 232 listings, followed by host_id 30283594 with 121 listings.

    o	The number of listings for the remaining hosts decreases gradually.

•	Distribution of Host Listings Count:

    o	The majority of hosts (around 32,303 hosts) have only 1 listing.

    o	The distribution of host listings count follows a decreasing trend, with fewer hosts having multiple listings. 

    o	Some hosts have a significant number of listings, as evidenced by the presence of hosts with 327, 232, 121, and 103 listings.

•	Host Statistics:

    o	These statistics provide insights into the average pricing and popularity of listings hosted by different hosts.
